```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style = "whitegrid", palette = "Set2")
```

## 1. Countplot

Similar to countplot, but shows the distribution of occurrences of certain values in our data.

```python
penguin = sns.load_dataset("penguins")
```

```python
penguin
```

```python
sns.countplot(data = penguin, x = "species")
```

```python
tit = sns.load_dataset("titanic")
```

```python
sns.countplot(data = tit, x = "sex", hue = "pclass")
```

```python
sns.countplot(data = tit, y = "sex", hue = "pclass")
```

## 2. Strip & Swarm Plots

```python
tip = sns.load_dataset("tips")
```

```python
sns.stripplot(data = tip, x = "tip", y = "day")
```

```python
sns.swarmplot(data = tip.nlargest(100,"total_bill"), x = "tip", y = "day")
```

The strip will overlap, but swarm will distribute averagely.

```python
sns.swarmplot(data = tit, x = "pclass", y = "age", hue = "sex")
```

## 3. boxplot

```python
sns.boxplot(data = tit, x = "age")
```

```python
sns.boxplot(data = tit,x = "sex",y = "age")
```

```python
sns.boxplot(data = tit,x = "sex",y = "age")
sns.swarmplot(data = tit, x = "sex", y = "age", alpha = .4)
```

## 5. Violinplot

A violin plot plays a similar role as a box-and-whisker plot. It shows the distribution of data points after grouping by one (or more) variables. Unlike a box plot, each violin is drawn using a kernel density estimate of the underlying distribution.

```python
sns.violinplot(data = tit, x = "age")
plt.figure()
sns.boxenplot(data = tit, x = "age")
```

```python
sns.violinplot(data = tit,x = "sex",y = "age")
```

```python
sns.violinplot(data = tit,x = "pclass",y = "age", hue = "sex")
```

```python
sns.violinplot(data = tit,x = "pclass",y = "age", hue = "sex", split = True)
```

## 6. barplot

A bar plot represents an aggregate or statistical estimate for a numeric variable with the height of each rectangle and indicates the **uncertainty** around that estimate using an **error bar**.

```python
sns.barplot(data = tit,x = "pclass",y = "age", hue = "sex")
```

Show us the mean value by default.

```python
sns.barplot(data = tit,x = "pclass",y = "age", hue = "sex", dodge = False)
```

```python
sns.barplot(data = tit,y = "pclass",x = "age", hue = "sex", orient = "h")
```

If switch `x` and `y` up and want it to be horizontal, it is not enough.
Because it can't figure out which one of these is categorical.
They're both numeric.
So we need to specifically say `orient`.

## 7.catplot

Now cat plot allows us to work on a figure and create a whole bunch of plots potentially at the figure level that are of these different types.

+ height
+ alpha
+ kind
+ ...

https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot

```python
sns.catplot(data = tit, kind = "bar",x = "sex", y = "survived", col = "pclass" )
```

```python
sns.catplot(data = tit, kind = "bar",x = "sex", y = "survived", col = "pclass",aspect = .5)
```
