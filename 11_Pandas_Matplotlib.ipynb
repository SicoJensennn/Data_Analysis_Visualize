```python
salaries = pd.read_csv("data/Salaries.csv", low_memory=False)
```

```python
salaries.head(3)
```

```python
salaries.info()
```

```python
salaries["BasePay"] = pd.to_numeric(salaries["BasePay"], errors="coerce").fillna(0)
salaries["OvertimePay"] = pd.to_numeric(salaries["OvertimePay"], errors="coerce").fillna(0)
salaries["OtherPay"] = pd.to_numeric(salaries["OtherPay"], errors="coerce").fillna(0)
```

```python
df = salaries[["EmployeeName", "BasePay", "OvertimePay", "OtherPay"]]
```

```python
df.set_index("EmployeeName", inplace = True)
```

```python
df.head(10).plot(kind = "bar")
```

Or I can simply use `plot.bar()`

```python
df["BasePay"].sort_values(ascending = False).head().plot.bar()
```

```python
df.sort_values("OvertimePay", ascending = False).head(6).plot.barh(stacked = True)
```

## 6. Scatter plot

```python
houses.plot.scatter(x = "bedrooms", y = "bathrooms", marker = "+",color = "#52796f")
```

```python
ufo[ufo["state"] == "UT"].year.value_counts().sort_index().plot(color = "#738290",label = "UT")
ufo[ufo["state"] == "OR"].year.value_counts().sort_index().plot(color = "#a1b5d8",label = "OR")
ufo[ufo["state"] == "CA"].year.value_counts().sort_index().plot(color = "#7ebc89",label = "CA")
ufo[ufo["state"] == "AZ"].year.value_counts().sort_index().plot(color = "#ffb627",label = "AZ")
plt.legend()
```

```python
ufo[ufo["shape"] == "light"].year.value_counts().sort_index().plot(color = "#738290", label = "light")
ufo[ufo["shape"] == "fireball"].year.value_counts().sort_index().plot(color = "#a1b5d8", label = "fireball")
ufo[ufo["shape"] == "circle"].year.value_counts().sort_index().plot(color = "#7ebc89",label = "circle")
ufo[ufo["shape"] == "triangle"].year.value_counts().sort_index().plot(color = "#f5cb5c",label = "triangle")
plt.xlim(2000, 2018)
plt.legend()
```

## Automatic Subplots

```python
df.plot(kind = "hist", subplots = True, sharex = False, figsize=(5,8))
```

```python
axs = df.plot(kind = "hist", subplots = True, sharex = False, layout = (1,3),figsize=(12,4), bins = 30)
axs[0][1].set_title("Overtimerpay")
axs[0][1].set_xlim(0,100000)
axs[0][2].set_title("Otherpay")
axs[0][2].set_xlim(0,80000)
axs[0][0].set_title("Basepay")
axs[0][0].set_xlim(0,250000)
```

```python
fig, axs = plt.subplots(1,2)
ufo.year.value_counts().sort_index().plot(ax=axs[0])
axs[0].set_title("Sightings By Year")
ufo.month.value_counts().sort_index().plot(kind="bar", ax=axs[1])
axs[1].set_title("Sightings By Month")
```

## 8. Export Fig

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html#matplotlib.pyplot.savefig

```python
fig, axs = plt.subplots(1,2)
ufo.year.value_counts().sort_index().plot(ax=axs[0])
axs[0].set_title("Sightings By Year")
ufo.month.value_counts().sort_index().plot(kind="bar", ax=axs[1])
axs[1].set_title("Sightings By Month")
plt.savefig("ufo")
```
