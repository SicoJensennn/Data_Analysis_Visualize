## 6. invert or negate boolean series 

```python
new_b = houses['yr_built'] >= 2014
new_r = houses['yr_renovated'] >= 2014
```

```python
recent = new_b | new_r
```

```python
old = houses[~recent]
```

```python
old
```

```python
tit
```

```python
alive = titanic.survived == 1
titanic[~alive]
```

```python
child = titanic.age <= "14"
children_alive = alive & child
```

```python
titanic[children_alive]
```

```python
titanic[~children_alive]  # age > 14, or died
```

注意和、或对调
+ `~ & = |`
+ `~ | = &`
