## Sort_values

```python
wh.sort_values("Ladder score")
```

Ascending values by default.

```python
wh.sort_values("Ladder score", ascending = False).head()
```

Sorting by 2 values. (in order)

```python
wh.sort_values(["Generosity","Freedom to make life choices"], ascending = False).head()
```

### sorting text 
order: ABCDEF…abcdef

```python
wh.sort_values(["Regional indicator"]).head()
```

Using `key`:callable, optional.
Apply the key function to the values before sorting.It should expect a Series and return a Series with the same shape as the input. It will be applied to each column in by independently.

```python
wh.sort_values(["Regional indicator"], key = lambda col: col.str.lower()).head()
# lowercase when making comparison. The data won't change.
```

### loc
It helps us view or access data by row rather than by column.
`.loc[]` is primarily label based, but may also be used with a boolean array.

```python
wh.loc["China"]
```

```python
wh.loc[["China"]]
```

It tells loc to give dataframe.

```python
wh.loc[["China","Canada","United States"]].sort_values("Freedom to make life choices")
```

```python
titanic.loc[[3]]
# by number
```

```python
titanic.loc[3:10:3]
# slice
```

```python
wh.sort_index().loc["China":"Denmark"]
```

```python
wh.sort_index().loc["Denmark":"China":-1]
```

### iloc
int location.

```python
titanic.iloc[2]
```

```python
titanic.sort_index(ascending=False, inplace=True)
```

```python
titanic.head()
```

```python
titanic.iloc[0:4]
```

`iloc` is based on position.
`loc` is based on values.

```python
titanic.loc[0:4:-1]
# 因为现在的排列顺序是从4到0.
```

### loc & iloc with Series

```python
titanic.loc[0:6:-1, ['pclass','sex', 'age']]
```

```python
titanic['age'].value_counts()
```

```python
titanic['age'].value_counts().loc['18']
```

```python
titanic['age'].value_counts().iloc[0:5]
```
