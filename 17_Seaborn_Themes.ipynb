```python
import seaborn as sns
```

+ white
+ dark
+ whitegrid
+ darkgrid
+ ticks

```python
tip = sns.load_dataset("tips")
```

```python
sns.set_style("white")
sns.scatterplot(data = tip, x = "total_bill", y = "tip")
```

```python
sns.set_style("ticks")
sns.scatterplot(data = tip, x = "total_bill", y = "tip")
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data = tip, x = "total_bill", y = "tip")
```

### What if I wanna change the font size?

```python
sns.axes_style()
```

Rewrite the setting:

```python
sns.set_style({'axes.facecolor': '#f7ede2',
               'grid.color': '#c9cba3'})
sns.scatterplot(data = tip, x = "total_bill", y = "tip")
```

## Despine

Remove the top and right spines form plots.

```python
sns.set_style("white")
```

```python
sns.histplot(data = tip, x = "total_bill")
sns.despine()
```

If I wanna remove bottom spine.

```python
sns.histplot(data = tip, x = "total_bill")
sns.despine(bottom = True)
```
