**Seaborn** is a data visualization library that is built on top of matplotlib.

+ provides a high-level interface.
+ makes it easy to create attractive charts out of the box
+ its plotting methods operate on dataframes.

https://seaborn.pydata.org/

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## 1. Load_dataset

Load an example dataset from the online repository (requires internet).

https://github.com/mwaskom/seaborn-data

https://seaborn.pydata.org/generated/seaborn.load_dataset.html#seaborn-load-dataset

```python
sns.load_dataset("tips")
```

## 2.scatterplot

```python
sns.set_theme()
```

```python
tip = sns.load_dataset("tips")
```

```python
sns.scatterplot(data = tip, x = "total_bill", y = "tip")

# data: tips
# x
# y
```

```python
tip.plot(kind = "scatter", x = "total_bill", y = "tip")
```

```python
sns.scatterplot(data = tip, x = "total_bill", y = "tip", hue ="smoker", style = "time")
```

`hue`: Grouping variable that will produce points with different colors. 

`style`: Grouping variable that will produce points with different markers. 

`size`: Grouping variable that will produce points with different sizes. 

```python
sns.scatterplot(data = tip, x = "total_bill", y = "tip", size = "size")
```

## 3. lineplots

https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn-lineplot

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

```python
sns.lineplot(data = flights, x = "year", y = "passengers")
```

Plotting the average number of passengers.

`estimator`: 
Method for aggregating across multiple observations of the y variable at the same x level. If None, all observations will be drawn. The default estimator is mean.

```python
sns.lineplot(data = flights, x = "year", y = "passengers", estimator = "sum")
```

```python
sns.lineplot(data = flights, x = "year", y = "passengers", estimator = "max")
```

```python
sns.lineplot(data = flights, x = "year", y = "passengers", hue = "month")
```

## 4. relplot

```python
sns.relplot(tip, kind = "scatter", x="total_bill",y = "tip", hue = "smoker", col = "time", row = "sex")
```

## Histplot

```python
sns.histplot(data = tip, x = "tip")
```

```python
sns.histplot(data = tip, x = "tip", hue = "smoker")
```

default: overlapping. stacked on each other.

```python
sns.histplot(data = tip, x = "tip", hue = "smoker", multiple = "stack")
```

```python
sns.histplot(data = tip, x = "tip", hue = "smoker", multiple = "dodge")
```

```python
sns.histplot(data = tip, x = "tip", hue = "smoker", binwidth = 10)
```

```python
sns.histplot(data = tip, x = "tip", hue = "smoker", bins = 10)
```

```python
sns.histplot(data = tip, x = "tip", hue = "smoker", element = "step")
```

`element`: Visual representation of the histogram statistic. Only relevant with univariate data.

`kde`: If True, compute a kernel density estimate to smooth the distribution and show on the plot as (one or more) line(s). 

```python
sns.histplot(data = tip, x = "tip", hue = "smoker",kde = True)
```

## 7. KDE plot

```python
sns.kdeplot(data = tip, x = "total_bill", hue = "day")
```

```python
sns.kdeplot(data = tip, x = "total_bill", hue = "day", bw_adjust = 0.5)
```

`bw_adjust`: Factor that multiplicatively scales the value chosen using bw_method. Increasing will make the curve smoother. 

```python
sns.kdeplot(data = tip, x = "total_bill", hue = "day", bw_adjust = 4)
```

```python
sns.kdeplot(data = tip, x = "total_bill", hue = "day", multiple = "stack")
```

## 8. bivariate distribution plots

```python
sns.histplot(data = tip, x = "tip", y = "total_bill")
```

You can see it as a heat plot. The darker, the higher density. (More frequently)

```python
sns.kdeplot(data = tip, x = "tip", y = "total_bill")
```

```python
sns.kdeplot(data = tip, x = "tip", y = "total_bill", cmap = "Blues",fill=True)
```

## 9. Rugplot

用于绘制出一维数组中数据点实际的分布位置情况，即不添加任何数学意义上的拟合，单纯的将记录值在坐标轴上表现出来

```python
sns.kdeplot(data = tip, x = "tip")
sns.rugplot(data = tip, x = "tip")
```

```python
sns.scatterplot(data = tip, x = "tip", y = "total_bill")
sns.rugplot(data = tip, x = "tip",y = "total_bill", alpha = .5)
```

Using alpha to change transparency.
