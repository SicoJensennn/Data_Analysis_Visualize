It will output the chart.

```python
import numpy as np
nums = np.arange(5)  # Return evenly spaced values within a given interval.
```

```python
plt.plot(nums,nums)
plt.plot(nums,nums*nums)
plt.plot(nums,nums**nums)
```

It didn't make a separate plot. Instead, it combines 2 plot together.

## 2. Figure

Figure is the top level thing.It includes couple of plots.
Any thing that comes after figure is added until we make a new figure.

```python
plt.figure()
plt.plot(nums,nums)
plt.figure()
plt.plot(nums,nums*nums)
```

## 3. control the size of figure
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html#matplotlib.pyplot.figure

`figsize(float, float)`, default: [6.4, 4.8] Width, height in inches.

```python
plt.figure(figsize = (12,4))
plt.plot(nums,nums**0.5)
```

`dpi`: control the number of pixel per inch.(default: 100.0)

```python
plt.figure(figsize = (12,4), dpi = 40)
plt.plot(nums,nums**0.5)
```

## 5. Title & Label

```python
ages = [19,22,28,31,45,62]
salaries = [1700,2300,2600,3300,3700,4600]
plt.plot(ages, salaries)
plt.title("Company Salaries", loc = "left")
plt.xlabel("ages", labelpad = 10)
plt.ylabel("average salaries($)", labelpad = 20)
```

**tick**


**lim**
The range of axis.

```python
plt.plot(ages, salaries)
plt.title("Company Salaries", loc = "left")
plt.xlabel("ages", labelpad = 10)
plt.ylabel("average salaries($)", labelpad = 20)
plt.xticks([10,20,30,40,50,60,70])
plt.ylim(1500,5000)
```

## 6. Add legend
https://matplotlib.org/stable/api/legend_api.html#module-matplotlib.legend

```python
spending = [1000,1200,1500,1600,2100,1700]
plt.plot(ages, salaries, label = "average salaries", color = "#b8b8ff")
plt.plot(ages, spending, label = "average cost", color = "#abc4ab")
plt.title("Company Salaries")
plt.xlabel("ages")

plt.xticks([10,20,30,40,50,60,70])
plt.ylim(500,5000)
plt.legend(loc = "lower right", frameon = False, shadow = False)
# lower, center, upper
```

## 7. Bar Plots

```python
plants = ['spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10,25,5,30,21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.bar(plants, germinated, width = 0.5,color = "#a3b18a")
plt.bar(plants, died, width = 0.5,color = "#d4a373", bottom = germinated, label = "died seeding")
plt.legend(frameon = False)
plt.title("Planting Seeds")
```

Horizontal.

```python
plt.barh(plants, germinated,color = "#bde0fe")
```

```python
import pandas as pd
titanic = pd.read_csv("data/titanic.csv")
titanic.age = pd.to_numeric(titanic["age"], errors = "coerce")
```

```python
titanic.age
```

```python
plt.hist(titanic.age)
```

```python
class_1 = titanic.pclass == 1
class_2 = titanic.pclass == 2
class_3 = titanic.pclass == 3
```

```python
plt.hist(titanic[class_1].age,color = "#bde0fe", label = "1st class", alpha = 0.6, bins = 15)
plt.hist(titanic[class_2].age,color = "#cdb4db", label = "2nd class", alpha = 0.6, bins = 15)
plt.hist(titanic[class_3].age,color = "#ffc8dd", label = "3rd class", alpha = 0.6, bins = 15)
plt.legend(frameon = False)
plt.title("Class Ages")
```

alpha: transparency.
